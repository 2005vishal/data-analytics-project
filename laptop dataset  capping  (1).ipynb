```python
def con_obj(y,x,z):## x is group feature and z is opertional
    d = y.groupby(x)[z].agg(['count','mean']).reset_index()
    plt.figure(figsize = (12,19))
    plt.subplot(2,1,1)
    ax = sns.barplot(x = d[x],y = d['mean'])
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
    plt.ylabel(z)
    plt.title(f'fequrency distributed  Average price on {x}')
    plt.subplot(2,1,2)
    ax = sns.barplot(x = d[x],y = d['count'])
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
    plt.ylabel('count')
    plt.title(f'fequrency distributed count on {x}')
    plt.show() 
    return d
    
    
```
