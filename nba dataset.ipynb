```python
l = list(df1['Height'])
d = {}
for i in range(len(l)):
    b = l[i].replace('-','.')
    d[b] = l[i]
```

```python
def hg_data(data):
    for k,v in d.items():
        if data == v:
            return float(k)
df1['hgt'] = df1.Height.apply(hg_data)
```

```python
df1.drop(columns = ['Height','College'],axis = 1,inplace = True)
```

I have check the outlier in the data. i will  remove the outlier in the datase

```python
q1 = np.percentile(df1['Salary'],25)
q3 = np.percentile(df1['Salary'],75)
iqr = q3 - q1

upper_sl = q3 + 1.5*iqr
lower_sl = q1 -  1.5*iqr
print("upper salary",upper_sl )
print("Lower salary",lower_sl)
```

```python
q1 = np.percentile(df1['Age'],25)
q3 = np.percentile(df1['Age'],75)
iqr = q3 - q1

upper_ag = q3 + 1.5*iqr
lower_ag = q1 -  1.5*iqr
print("upper age",upper_ag)
print("Lower age",lower_ag)
```

```python
q1 = np.percentile(df1['Weight'],25)
q3 = np.percentile(df1['Weight'],75)
iqr = q3 - q1

upper_wg = q3 + 1.5*iqr
lower_wg = q1 -  1.5*iqr
print("upper weight",upper_wg)
print("Lower weight",lower_wg)
```

```python
q1 = np.percentile(df1['Number'],25)
q3 = np.percentile(df1['Number'],75)
iqr = q3 - q1

upper_nm = q3 + 1.5*iqr
lower_nm = q1 -  1.5*iqr
print("upper number",upper_nm)
print("Lower number",lower_nm)
```

```python
q1 = np.percentile(df1['hgt'],25)
q3 = np.percentile(df1['hgt'],75)
iqr = q3 - q1

upper_hg = q3 + 1.5*iqr
lower_hg = q1 -  1.5*iqr
print("upper Height",upper_hg)
print("Lower Height",lower_hg)
```

```python
df = df1[(df1["Salary"] < upper_sl) & (df1['Age'] > lower_ag) & (df1['Age'] < upper_ag) & (df1['Number'] < upper_nm) & (df1['Weight'] < upper_wg) & (df1['Weight'] > lower_wg) & (df1['hgt']>lower_hg) & (df1['hgt'] < upper_hg)]
```

```python
def age_cat(data):
    if data in range(19,26): return "Younger Player"
    elif data in range(26,32): return "Experience Player" 
    else: return "Senior Player"
df['Age_category'] = df['Age'].apply(age_cat)
```

```python
# df.drop(columns = 'Age_category',axis = 1,inplace = True)
```

```python
sal = df.groupby(df['Age_category'])['Salary'].mean().reset_index()
ax = sns.barplot(x = 'Age_category',y = 'Salary' , data = sal,palette = list('rgb'))
for i in range(sal.Salary.nunique()):
    ax.bar_label(ax.containers[i]) 
plt.show()
```

```python
sal = df.groupby(df['Age_category'])['Weight'].mean().reset_index()
ax = sns.barplot(x = 'Age_category',y = 'Weight', data = sal,palette = list('rgb'))
for i in range(sal.Weight.nunique()):
    ax.bar_label(ax.containers[i])
plt.title('Average Weight Age Category')
plt.show()
```

```python
sal = df.groupby(df['Age_category'])['hgt'].mean().reset_index()
ax = sns.barplot(x = 'Age_category',y = 'hgt', data = sal,palette = list('rgb'))
for i in range(sal.hgt.nunique()):
    ax.bar_label(ax.containers[i])
plt.title('Average Height Age Category')
plt.show()
```

```python
def lbs_to_kg(lbs):
    return lbs / 2.20462
df['Weight_kg'] = df.Weight.apply(lbs_to_kg)
```

```python
wei = round(df.groupby(df['Position'])['Weight_kg'].mean().reset_index())
ax = sns.barplot(x = wei['Position'],y = wei['Weight_kg'],palette = list('rgb'))
for i in range(wei.Weight_kg.nunique()):
    ax.bar_label(ax.containers[i])
plt.title('Avgerage Weight by position')
plt.show()
```

```python
wei = round(df.groupby(df['Position'])['hgt'].mean().reset_index())
ax = sns.barplot(x = wei['Position'],y = wei['hgt'],palette = list('rgb'))
for i in range(len(wei.hgt)):
    ax.bar_label(ax.containers[i])
plt.title('Avgerage Height by position (in feet)')
plt.show()
```
